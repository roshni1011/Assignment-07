```python
#!pip install mlxtend
import pandas as pd
from sklearn.preprocessing import normalize
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
EastWestAirlines_data = pd.read_excel('EastWestAirlines.xlsx')
EastWestAirlines_data
```

```python
EastWestAirlines_data1=EastWestAirlines_data.drop(['ID#'],axis=1)
EastWestAirlines_data1
```

```python
# Normalize heterogenous numerical data 
EastWestAirlines_data1_norm=pd.DataFrame(normalize(EastWestAirlines_data1),columns=EastWestAirlines_data1.columns)
EastWestAirlines_data1_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(EastWestAirlines_data1_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(EastWestAirlines_data1_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
EastWestAirlines_data1['clustersid']=hclusters.labels_
EastWestAirlines_data1
```

```python
EastWestAirlines_data1.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(EastWestAirlines_data1['clustersid'],EastWestAirlines_data1['Balance'], c=hclusters.labels_) 
```
